```python
import numpy as np
import matplotlib.pyplot as plt

pupil = pupil.loc[pupil['confidence'] > 0.99]
left_pupil_time = pupil.loc[pupil['eye_id'] == 0]['timestamp']
right_pupil_time = pupil.loc[pupil['eye_id'] == 1]['timestamp']
left_pupil = pupil.loc[pupil['eye_id'] == 0][['timestamp', 'norm_pos_x', 'norm_pos_y']]
right_pupil = pupil.loc[pupil['eye_id'] == 1][['timestamp', 'norm_pos_x', 'norm_pos_y']]
```

```python
import matplotlib.pyplot as plt

plt.plot(left_pupil_time, left_pupil[['norm_pos_x', 'norm_pos_y']])
plt.legend(['norm_pos_x', 'norm_pos_y'])
```

```python
def detect_saccades(pupil_positions, velocity_threshold=1, min_duration=0.05):

    pupil_positions = pupil_positions.reset_index(drop=True)
    pupil_positions['velocity_x'] = np.abs(pupil_positions['norm_pos_x'].diff()) / pupil_positions['timestamp'].diff()
    pupil_positions['velocity_y'] = np.abs(pupil_positions['norm_pos_y'].diff()) / pupil_positions['timestamp'].diff()

    pupil_positions.loc[(pupil_positions['velocity_x'] > velocity_threshold) | (pupil_positions['velocity_y'] > velocity_threshold), 'velocity_threshold_surpassed'] = True
    pupil_positions['velocity_threshold_surpassed'] = pupil_positions['velocity_threshold_surpassed'].fillna(False)

    saccades = list()
    i = 0

    while i < len(pupil_positions.index):
        if pupil_positions.loc[i, 'velocity_threshold_surpassed']:
            start_time = pupil_positions.loc[i, 'timestamp']
            end_time = pupil_positions.loc[i, 'timestamp']

            while pupil_positions.loc[i, 'velocity_threshold_surpassed'] and i < len(pupil_positions.index):
                i += 1
                end_time = pupil_positions.loc[i, 'timestamp']

            if end_time - start_time > min_duration:
                saccades.append((start_time, end_time))
        i += 1

    return saccades
```

```python
saccades = detect_saccades(left_pupil, velocity_threshold=0.6, min_duration=0.05)
```

```python
# Plot the pupil position data and mark detected saccades
plt.plot(left_pupil['timestamp'], left_pupil[['norm_pos_x', 'norm_pos_y']], label='Pupil Position')
for saccade in saccades:
    plt.axvspan(saccade[0], saccade[1], facecolor='none', edgecolor='red', label='Saccade')
plt.xlabel('Time (s)')
plt.ylabel('Pupil Position')
plt.show()
```

```python
plt.plot(left_pupil['timestamp'].iloc[600:1000], left_pupil[['norm_pos_x', 'norm_pos_y']].iloc[600:1000], label='Pupil Position')
for saccade in saccades:
    if saccade[0] in left_pupil['timestamp'].iloc[600:1000].tolist():
        plt.axvspan(saccade[0], saccade[1], color='red', alpha = 0.2, label='Saccade')
plt.xlabel('Time (s)')
plt.ylabel('Pupil Position')
plt.show()
```
